## Data Collection using SQL command

Data was downloaded from the database using the below SQL command. The city chosen is New York city

```SQL
SELECT c.year, c.city, c.country, c.avg_temp, g.avg_temp AS global_avg_temp
FROM city_data AS c
INNER JOIN global_data AS g
ON c.year = g.year
WHERE c.city = 'New York';
```
The above command helps in accumulating the entire global and New York city trends in a single csv file. As the data was clean no more steps were necessary for cleaning the data.
