## Data Analysis and Visualization

The data present in the twitter_archive_master.csv has about 1132 rows and 23 columns. The insights I have drawn from the data are

1. Top 10 dog breeds tweeted by WeRateDogs


2. The relation between the Retweeted tweet and Favorite tweet changed over the years


3. Rating variation concerning the retweet count

This gives more insight into the type of data we are dealing with. How the WeRate Dogs account has evolved over the years in terms of ratings and the type of dogs they more commonly tweet about and has the account raised popularity in terms of retweeting and favorite counts over the years.  

1. The Golden retriever was the most commonly rated dog on the WeReteDogs account. Pembroke is the second most and Labrador Retriever being the third top most rated dog.

2. The favorite count has increased over time slowly but the retweet count has been consistent with the frequency over the years.

3. The rating has been consistent with the retweet count. A few tweets however with higher ratings have been retweeted more times
